### Exercise 2: Bar Plot with Colors

```python
# TODO: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
```

```python
import matplotlib.pyplot as plt
```

```python

# Define the categories
categories = ['A', 'B', 'C', 'D']

# Define the values
values =[5, 7, 3, 9]

# Create the bar plot with different colors
plt.bar(categories, values, color=['red', 'blue', 'green', 'purple'])

# Add a title to the plot
plt.title('Bar Plot with Different Colors')

# Display the plot
plt.show()
```
